Listings Details

```python
#Column Names
listings.columns
```

```python
#Data Types
listings.dtypes
```

```python
#Missing Values
listings.isnull().sum()
```

```python
missing = listings.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
#Duplicate Rows
listings.duplicated().sum()
```

```python
#Summary Statistics
listings.describe()
```

```python
#Dataset Information
listings.info()
```

```python
#Create Data Quality Summary
quality = pd.DataFrame({
    "Rows":[listings.shape[0]],
    "Columns":[listings.shape[1]],
    "Missing":[listings.isnull().sum().sum()],
    "Duplicates":[listings.duplicated().sum()]
})

quality
```
